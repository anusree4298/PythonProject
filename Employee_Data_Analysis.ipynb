# **Analysis Tasks:**
**1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.**

```python
#finding the distribution of employees across each team
team_distribution = df['Team'].value_counts()
print("The distribution of employees across each team:\n",team_distribution,"\n")

#calculating percentage split relative to total number of employees
total_employees=len(df)
percentage_split=(team_distribution/total_employees)*100
print("percentage split relative to total number of employees:\n",percentage_split)

#visualization
plt.figure(figsize=(8, 4))
sns.barplot(x=team_distribution.index, y=team_distribution.values)
plt.xlabel("Team")
plt.ylabel("Percentage of Employees")
plt.title("Employee Distribution Across Teams")
plt.xticks(rotation=90)
plt.show()
```

**2. Segregate employees based on their positions within the company.**

```python
position_distribution = df['Position'].value_counts()
print("segregating based on position:\n",team_distribution,"\n")

plt.pie(position_distribution, labels=position_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Segregating based on positions')
plt.show()
```

**3. Identify the predominant age group among employees.**

```python
df["Age Group"] = pd.cut(df["Age"], bins=[20, 30, 40, 50], labels=["20-30", "31-40", "41-50"],right=False)
age_group_distribution = df["Age Group"].value_counts()
print(age_group_distribution)

#visualization
plt.figure(figsize=(8, 4))
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values,color="red")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Age groups")
plt.xticks(rotation=90)
plt.show()
```

**4. Discover which team and position have the highest salary expenditure.**


```python
team_salary = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)
print(team_salary)

position_salary = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)
print(position_salary)

# Visualization for Teams
plt.figure(figsize=(10, 5))
plt.plot(team_salary.index, team_salary.values, color='yellow',marker='o')
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Team")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Visualization for Positions
plt.figure(figsize=(5, 3))
sns.barplot(x=position_salary.index, y=position_salary.values, color="magenta")
plt.xlabel("Position")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Position")
plt.xticks(rotation=90)
plt.show()
```

**5. Investigate if there's any correlation between age and salary, and represent it visually.**


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Age"], y=df["Salary"], alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation Between Age and Salary")
plt.show()

# Correlation value
correlation = df["Age"].corr(df["Salary"])
print(f"Correlation between Age and Salary: {correlation:.2f}")
```
